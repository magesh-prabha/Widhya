```python
import pandas as pd
```

```python
from sklearn import linear_model
```

```python
data=pd.read_excel('F:\\Projects\\Widhya-Bitcoin_Price_Prediction.xlsx')
data.loc[1023]['btc_market_price']
```

```python
import seaborn as sns
```

```python
df=pd.DataFrame(data)
corr=df.corr()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize = (12, 7))
sns.heatmap(corr,annot=True)
```

```python
model=linear_model.LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df=df.fillna(df.mean())
```

```python
X=df[['btc_market_cap','btc_n_transactions','btc_miners_revenue','btc_cost_per_transaction','btc_difficulty','btc_hash_rate','btc_cost_per_transaction_percent']]
y=df['btc_market_price']
```

```python
X_train,X_valid,y_train,y_valid=train_test_split(X,y)
```

```python
model.fit(X_train,y_train)
```

```python
predict=model.predict(X_valid)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_valid,predict)
```
